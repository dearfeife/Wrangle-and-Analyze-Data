### Quality

### Define
<p> None and NaN are not consistency, so we need to change those columns:doggo,floofer,pupper,puppo with 'None' to 'NaN'</p>

### Code

```python
name = 'None'
doggo = twitter_new['doggo']
floofer = twitter_new['floofer']
pupper = twitter_new['pupper']
puppo = twitter_new['puppo']
name_list = [doggo,floofer,pupper,puppo]
for  name in name_list:
    name.replace(name, np.nan,inplace = True)
```

### Test

```python
twitter_new
```

### Define
an,a,None used in name column, change those to NaN

### Code

```python
twitter_new.name = twitter_new.name.astype(str)
wrong_list = ['an','a','None']
for i in wrong_list:
    twitter_new.name = twitter_new.name.replace(i, np.nan)
```

### Test

```python
twitter_new['name']
```

### Define
tweet_id,in_reply_to_status_id,in_reply_to_user_id,retweeted_status_id,retweeted_status_user_id type need to change to string.

### Code

```python
# change NaN to 0, and then change the column to string 
twitter_new['tweet_id'] = twitter_new['tweet_id'].astype(str)
twitter_new['in_reply_to_status_id'].replace(np.nan,0,inplace = True)
twitter_new['in_reply_to_status_id'] = twitter_new['in_reply_to_status_id'].astype(int).astype(str)

twitter_new['in_reply_to_user_id'].replace(np.nan,0,inplace = True)
twitter_new['in_reply_to_user_id'] = twitter_new['in_reply_to_status_id'].astype(int).astype(str)

twitter_new['retweeted_status_id'].replace(np.nan,0,inplace = True)
twitter_new['retweeted_status_id'] = twitter_new['retweeted_status_id'].astype(int).astype(str)

twitter_new['retweeted_status_user_id'].replace(np.nan,0,inplace = True)
twitter_new['retweeted_status_user_id']=twitter_new['retweeted_status_user_id'].astype(int).astype(str)
#change 0 back to NaN
cols = ['in_reply_to_status_id', 'in_reply_to_user_id','retweeted_status_id','retweeted_status_user_id']
twitter_new[cols] = twitter_new[cols].replace({'0':np.nan, 0:np.nan})
twitter_new.sample(10)
```

### Define 
Invalid rating_nominator, need to extract from the text,and set values that greater than 10 to 11.

```python
twitter_new.rating_numerator = twitter_new.text.str.extract("([0-9]+[.]*[0-9]*)\/[0-9]+", expand = True)
twitter_new['rating_numerator']=twitter_new['rating_numerator'].astype(float)
twitter_new.loc[twitter_new.rating_numerator > 10, 'rating_numerator'] = 11
```

### Tidiness
### Define
<p>Extract each stage from the text column using regular expressions and pandas'str.extract method, and put into a new column called stages. Drop the 4 columns.

### Code

```python
twitter_new['stages'] = twitter_new.text.str.extract('(doggo|floofer|pupper|puppo)', expand = True)
twitter_new.drop(['doggo','floofer','pupper','puppo'], axis=1, inplace = True)
```

### Define
Merge the image_prediction and df_json by 'tweet_id'. Cancat the table and df_twitter table together

### Code

```python
new_table2 = image_prediction_new.merge(df_json_new,on ='tweet_id',how = 'left')
new_table3 = pd.concat([twitter_new,new_table2],axis = 1, join ='outer')
```
